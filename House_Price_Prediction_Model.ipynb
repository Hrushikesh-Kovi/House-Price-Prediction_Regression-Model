# ***Data Cleaning***

```python
df["date"] = pd.to_datetime(df["date"])
df["date"].dtypes
```

```python
new_column_names = [
    "Date",
    "Price",
    "Bedrooms",
    "Bathrooms",
    "SqftLiving",
    "SqftLot",
    "Floors",
    "WaterFront",
    "View",
    "Condition",
    "SqftAbove",
    "SqftBasement",
    "YearBuilt",
    "YearRenovated",
    "Street",
    "City",
    "StateZip",
    "Country"
]
df.columns = new_column_names
df.columns
```

# ***Data Preprocessing***

```python
df.hist(figsize=(18, 12))
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Date")
plt.title("Date Distribution")
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Date", kde=True)
plt.title("Date Distribution")
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Price")
plt.title("Original Price Distribution")
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Price", kde=True)
plt.title("Original Price Distribution")
plt.tight_layout()
```

```python
Q1, Q3 = df["Price"].quantile(0.25), df["Price"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[df["Price"] <= upper]
np.sort(df["Price"].unique())
```

```python
df = df[df["Price"] >= 80000]
df = df.reset_index(drop=True)
df.shape
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Price")
plt.title("Processed Price Distribution")
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Price", kde=True)
plt.title("Processed Price Distribution")
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Bedrooms")
plt.title("Original Bedrooms Distribution")
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Bedrooms", kde=True)
plt.title("Original Bedrooms Distribution")
plt.tight_layout()
```

```python
Q1, Q3 = df["Bedrooms"].quantile(0.25), df["Bedrooms"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df["Bedrooms"] = df["Bedrooms"].clip(lower, upper)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Bedrooms")
plt.title("Processed Bedrooms Distribution")
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Bedrooms", kde=True)
plt.title("Processed Bedrooms Distribution")
plt.tight_layout()
```
